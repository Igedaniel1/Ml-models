```python
def show_distribution(var_data):
    min_val = var_data.min()
    max_val = var_data.max()
    mean_val = var_data.mean() 
    med_val = var_data.median()
    mod_val = var_data.mode()[0]

    print(var_data.name, '\nMinimum: {:.2f}\nMaximum: {:.2f}\nMean: {:.2f}\nMedian: {:.2f}\nMode: {:.2f}'.format(min_val, max_val, mean_val, med_val, mod_val))

    fig, ax = plt.subplots(2, 1, figsize=(10, 4))
    ax[0].hist(var_data)
    ax[0].set_ylabel('Frequency')

    ax[0].axvline(x=min_val, color='gray', linestyle='dashed', linewidth=1)
    ax[0].axvline(x=max_val, color='gray', linestyle='dashed', linewidth=1)
    ax[0].axvline(x=mean_val, color='cyan', linestyle='dashed', linewidth=1)
    ax[0].axvline(x=med_val, color='red', linestyle='dashed', linewidth=1)
    ax[0].axvline(x=mod_val, color='yellow', linestyle='dashed', linewidth=1)

    ax[1].boxplot(var_data, vert=False)
    ax[1].set_xlabel('value')
    fig.suptitle(var_data.name)
    fig.show()

delayFields = ['DepDelay', 'ArrDelay']
for col in delayFields:
    show_distribution(df_flights[col])
```

```python
ArrDelay_01pcentile = df_flights.ArrDelay.quantile(0.01)
ArrDelay_90pcentile = df_flights.ArrDelay.quantile(0.90)
df_flights = df_flights[df_flights.ArrDelay > ArrDelay_01pcentile]
df_flights = df_flights[df_flights.ArrDelay < ArrDelay_90pcentile]

DepDelay_01pcentile = df_flights.DepDelay.quantile(0.01)
DepDelay_90pcentile = df_flights.DepDelay.quantile(0.90)
df_flights = df_flights[df_flights.DepDelay > DepDelay_01pcentile]
df_flights = df_flights[df_flights.DepDelay < DepDelay_90pcentile]

for col in delayFields:
    show_distribution(df_flights[col])
```
