```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets, model_selection
```

```python
linnerud_data = datasets.load_linnerud()
```

```python
print(linnerud_data.target_names, linnerud_data.feature_names)
```

```python
Jumps_feature = linnerud_data.data[:, 2]
weight_target = linnerud_data.target[:, 0]
```

```python
Jumps_feature = Jumps_feature.reshape(-1, 1)
print(Jumps_feature.shape)
print(weight_target.shape)
print(Jumps_feature)
print(weight_target)
```

```python
Jumps_feature_train, Jumps_feature_test, weight_target_train, weight_target_test = model_selection.train_test_split(Jumps_feature, weight_target, test_size=0.2, random_state=42)
```

```python
model = linear_model.LinearRegression()
model.fit(Jumps_feature_train, weight_target_train)
```

```python
weight_test_prediction = model.predict(Jumps_feature_test)
```

```python
plt.scatter(Jumps_feature_test, weight_target_test, color='black')
plt.plot(Jumps_feature_test, weight_test_prediction, linewidth=3)
plt.xlabel('jumps feature')
plt.ylabel('weight')
plt.title('plot showing the weight against jumps done')
plt.show()
```
